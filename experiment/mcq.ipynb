```python
import os

# First, verify the file exists
env_path = r'C:\Users\brian\mcqgen\.env'
print(f"File exists: {os.path.exists(env_path)}")
print(f"Is a file: {os.path.isfile(env_path)}")
print(f"File size: {os.path.getsize(env_path) if os.path.exists(env_path) else 'N/A'} bytes")

# Try to read the file contents
try:
    with open(env_path, 'r', encoding='utf-8') as f:
        content = f.read()
        print(f"\n--- File Contents ---")
        print(repr(content))  # Shows hidden characters
        print(f"\n--- Readable Format ---")
        print(content)
except Exception as e:
    print(f"Error reading file: {e}")

# Try loading with verbose mode
from dotenv import load_dotenv
result = load_dotenv(dotenv_path=env_path, verbose=True, override=True)
print(f"\nLoad result: {result}")

# Check if variables are actually loaded despite False return
print(f"\nChecking environment variables:")
for key in ['OPENAI_API_KEY', 'API_KEY']:  # adjust to your actual variable names
    value = os.getenv(key)
    if value:
        print(f"{key}: {value[:10]}... (hidden)")
    else:
        print(f"{key}: Not found")
```
